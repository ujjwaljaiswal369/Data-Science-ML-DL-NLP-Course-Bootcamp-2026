```python
# Simple Calculator

num1 = float(input('Enter the first number'))
operation = input('Enter the operator')
num2 = float(input('Enter the second number'))

if operation == '+' :
    print(f'Result : {num1+num2}')
elif operation == '-':
    print(f'Result : {num1-num2}')
elif operation == '/' :
    if num2 !=0 :
        print(f'Result : {num1/num2}')
    else :
        print('Error')
elif operation == '*':
    print(f'Result : {num1*num2}')
```
