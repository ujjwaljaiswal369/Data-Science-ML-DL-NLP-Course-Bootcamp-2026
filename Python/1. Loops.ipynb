#### Loops

1. for Loop
   - Iterating over a range
   - Iterating over a string

2. while Loop
3. Loop Control Statements
    - break
    - continue
    - pass
4. Nested Loops

```python
range(5)
```

```python
## for loop
for i in range(5):
    print(i)
```

```python
for i in range(0,5,2):
    print(i)
```

```python
for i in range(5,-15,-2):
    print(i)
```

```python
### strings

str = 'Krish Naik'
for i in str:
    print(i)
```

```python
### While LOOP
# The while loop continues to execute as long as the condition is True.

count = 0

while count<5:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## break
## The break statement exits the loop permaturely

## break sstatement


for i in range(110):
    if i==21:
        break
    print(i)
```

```python
## continue

## The continue statement skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
# The pass statement is a null operation; it does nothing.

for i in range(10):
    if i==5:
        pass
    print(i)
```

```python
## Nested loopss
## a loop inside a loop

for i in range(5):
    for j in range(3):
        print(f'i:{i} and j:{j}')
```

```python
## Examples- Calculate the sum of first N natural numbers using a while and for loop

#for_loop

n= int(input())
sum_n = 0

for i in range(1,n+1):
    sum_n += i

print(f'The sum of first natural no. of {n} is {sum_n}')
```

```python
#while loop

n= int(input())
sum_n = 0
i=1
while i<=n:
    sum_n += i
    i += 1
print(sum_n)
```

```python
## Example- Prime numbers between 1 and 100

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        
        else:
            print(num)
```

#### Conclusion:
Loops are powerful constructs in Python that allow you to execute a block of code multiple times. By understanding and using for and while loops, along with loop control statements like break, continue, and pass, you can handle a wide range of programming tasks efficiently.
