**Question 5:** Write a Python program to calculate the factorial of a number.

```python
#function to find factorial of any no.
def factorial(num):
    result = 1
    for i in range(1,num+1):
         result *= i
    print(result)
```

```python
factorial(6)
```

```python
factorial(11)
```

**Question 12:** Write a Python program to demonstrate comparison operators: equal to, not equal to, greater than, less than.

```python
#Function_for_calculating_comparison_operations

def comparison_calculator(x,y):
    if (x==y):
        print('Numbers are equal')

    elif x!=y:
        print('Numbers are not equal')

        if x>y:
            print(f'{x} is greater than {y}')

        elif x<y:
            print(f'{y} is greater than {x}')
```

```python
#testing
comparison_calculator(1,2)
```

```python
comparison_calculator(1,1)
```

```python
comparison_calculator(23,100)
```

```python
comparison_calculator(123,324)
```

```python
comparison_calculator(45,54)
```

**Question 13:** Write a Python program to demonstrate logical operators: and, or, not.

```python
#Logical_Operators
a = 20
b = 10
c = 5

#AND operator return true only if both condition is true
print('AND operator')
print(a>b and b>c) 
print(b>a and b>c) #False because one of condition is not true

#OR operator return true only when one of given condition is true
print('OR operator')
print(a>b or b<c) 
print(a<b or b<c) #false because both condition is not true

#NOT operator return the oposite result if condition is true it return false
print('NOT operator')
print(not a>b)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
#Solution
year = int(input())

if (year%4 == 0 and year%100 != 0) or year%400 == 0:
    print('Its a Leap year')

else:
    print('Its a not a Leap year')
```
