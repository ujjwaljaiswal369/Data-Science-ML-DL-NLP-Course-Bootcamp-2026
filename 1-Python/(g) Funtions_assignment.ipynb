### Assignment 1: Fibonacci Sequence with Memoization

Define a recursive function to calculate the nth Fibonacci number using memoization. Test the function with different inputs.


```python
#solution-1
memo = {}
def fibonacci(n):
    '''Function to calculate fibonacci no. of given no.'''
    #basic_case
    for i in range(0,2):
        if n==i:
            return n
    
    #memoization
    if n in memo:
        return memo[n]
    
    #recursive formula
    memo[n] = fibonacci(n-1) + fibonacci(n-2)
    
    return memo[n]
    
```

```python
print(fibonacci(5),
      fibonacci(10),
      fibonacci(20),
      fibonacci(3),sep='\n')
```

### Assignment 3: Function with Variable Keyword Arguments

Define a function that takes a variable number of keyword arguments and returns a dictionary containing only those key-value pairs where the value is an integer. Test the function with different inputs.

```python
#solution-3

def int_dict(**args):
    result={}
    for key,value in args.items():
        if isinstance(value,int):
            result[key] = value
    return result
```

```python
print(int_dict(name='ujjwal', age=22, height=6), int_dict(height=12,width=21,item='axe'),sep='\n')
```

### Assignment 4: Function with Callback

Define a function that takes another function as a callback and a list of integers. The function should apply the callback to each integer in the list and return a new list with the results. Test with different callback functions.

```python
#solution-4

#Callback_functions
def square(x):
    return x**2

def cube(x):
    return x**3

#High-order function
def apply_order_function(list, callback_function):
    result=[]

    for num in list:
        result.append(callback_function(num))
    
    return result
```

### Assignment 6: Function with Decorators

Define a function that calculates the time taken to execute another function. Apply this decorator to a function that performs a complex calculation. Test the decorated function with different inputs.

```python
#solution-6
import time

#time_calculating_function
def time_calculating(func):
    
    def wrapper(n):
        start = time.time()
        result = func(n)
        end = time.time()
        
        print(f'Time taken : {end-start} seconds')
        return result
        
    return wrapper


#complex_calculation_function
@time_calculating
def complex_calculation(n):
    
    total = 0
    for i in range(n):
        total+= i*i
        
    return total
```

```python
#testing
print(complex_calculation(10000),
complex_calculation(50000),sep='\n')
```

### Assignment 7: Higher-Order Function for Filtering and Mapping

Define a higher-order function that takes two functions, a filter function and a map function, along with a list of integers. The higher-order function should first filter the integers using the filter function and then apply the map function to the filtered integers. Test with different filter and map functions.


```python
#solution-7

#filter_functions
def is_even(n):
    return n%2==0

def is_positive(n):
    return n>0

#map_functions
def square(n):
    return n**2
def twice(n):
    return n*n

#Higher_order_function
def higher_order_function(filter_func,map_func,num):
    
    result = []
    for i in num:
        if filter_func(i):
            result.append(map_func(i))
    
    return result
        
```

```python
#testing
lst = [1,2,3,0,-1,-2,-3,4,5]

print(higher_order_function(is_positive,square,lst),
      higher_order_function(is_even,twice,lst)
      ,sep='\n')
```

### Assignment 8: Function Composition

Define a function that composes two functions, f and g, such that the result is f(g(x)). Test with different functions f and g.


```python
#solution-8

#function-g
def add_two(n):
    return n + 2

#function-f
def square(n):
    return n**2

#function-composition
def function_composition(f,g,l):
    
    new_l = []
    for i in l:
        new_l.append(f(g(i)))
    
    return new_l
```

```python
#testing
lst = [1,2,3,4,5]

print(function_composition(add_two,square,lst))
```

### Assignment 10: Function with Error Handling

Define a function that takes a list of integers and returns their average. The function should handle any errors that occur (e.g., empty list) and return None in such cases. Test with different inputs.

```python
#solution-10

def average_calculate(lst):
    '''This function calculating the average of list of integers.'''
    
    #check_empty_list_or_not
    if lst==None or len(lst)==0:
        return None
    
    #check_invalid_datatype
    for i in lst:
        if type(i)==str or type(i)==bool or type(i)==complex:
            return None
     
    #averaging   
    average = sum(lst)/len(lst)
    
    print(average)
```

### Assignment 11: Function with Generators

Define a function that generates an infinite sequence of Fibonacci numbers. Test by printing the first 10 numbers in the sequence.


```python
#solution-11


def fibonacci():
    '''This function produces infinite sequence of fibonacci numbers'''
    a,b = 0,1
    
    while True:
        yield a
        a,b = b,a+b

#Creating_generator_object       
fib = fibonacci()
```

```python
#testing
for i in range(10):
    print(next(fib))
```

### Assignment 12: Currying

Define a curried function that takes three arguments, one at a time, and returns their product. Test the function by providing arguments one at a time.


```python
#solution-12

def curried_multiply(x):
    '''This function return the product of three input numbers.'''
    def next_step(y):
        
        def next(z):
            
            return x*y*z
    
        return next
    
    return next_step
```

```python
#testing
print(curried_multiply(1)(2)(3),

      curried_multiply(12)(12)(12),
sep='\n')
```

### Assignment 13: Function with Context Manager

Define a function that uses a context manager to write a list of integers to a file. The function should handle any errors that occur during file operations. Test with different lists.


```python
#solution-13

def list_of_int_to_file(list):
    '''This function is used for converting list of numbers to a file'''
    try:
        with open('Numbers.txt', 'w') as file:
            
            for i in list:
                file.write(str(i) + '\n')
            
        print('Data Written Successfully')
        
        
    except Exception as e:
        
        print('Error:',e)
            
    
```

```python
#testing
list_of_int_to_file([0,1,1,2,3,4,5,5,6,7,5])
```

### Assignment 14: Function with Multiple Return Types

Define a function that takes a list of mixed data types (integers, strings, and floats) and returns three lists: one containing all the integers, one containing all the strings, and one containing all the floats. Test with different inputs.


```python
#solution-14

def separating_datatypes(lst):
    
    integers = []
    strings = []
    float_no = []
    boolean = []
    
    for i in lst:
        if type(i)==int:
            integers.append(i)
        elif type(i)==str:
            strings.append(i)
        elif type(i)==float:
            float_no.append(i)
        elif type(i)==bool:
            boolean.append(i)
            
    return integers,strings,float_no,boolean
```

```python
list6 = [1,2,3,'hello','Data Science', 'ML', 1.2,2.3]
list7 = [103,2.5,740.0,'ujjwal21', 'thakurji11', 'qwerty', 12,23,34,True , False, None]
print(separating_datatypes(list6),
      separating_datatypes(list7),
      sep='\n')
```
