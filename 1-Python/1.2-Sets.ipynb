## Sets

Sets are a built-in data type in Python 
- sets is used to store collections of unique items. 
- They are unordered, meaning that the elements do not follow a specific order.
- They do not allow duplicate elements. 
- Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

```python
# create a set
my_set = {1,2,3,4,5}
my_set
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_set = set([1,2,3,4,5])
my_set
```

```python
my_set = set([1,2,32,6,5,4,5])
my_set
```

```python
## Basics Sets Operation
## Adding and Removing Elements
my_set
```

```python
my_set.add(33)
my_set
```

```python
my_set.remove(5)
my_set
```

```python
my_set.discard(33)
my_set
```

```python
# pop method
last_removed_element=my_set.pop()
last_removed_element
my_set
```

```python
#clear all the elements
my_set.clear()
my_set
```

```python
## Set Memebership test
my_set={1,2,3,4,5}
print(2 in my_set)
print(6 in my_set)
```

```python
#Mathmatical Operation (Union & Intersection)
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

union_set = set1.union(set2)
union_set
```

```python
intersection_set = set1.intersection(set2)
intersection_set
```

```python
set1.intersection_update(set2)
set1
```

```python
# Difference
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
#Symmetric_difference
set1.symmetric_difference(set2)
```

```python
set2.symmetric_difference(set1)
```

```python

#Sets Methods
set1 = {1,2,3,4,5}
set2 = {3,4,5}

set1.issubset(set2)
```

```python
set1.issuperset(set2)
```

```python
lst = [1,2,3,3,4,5,6,6]

set(lst)
```

```python
## Counting unique words in text
text = 'This files contains code of learning sets. Sets are very useful'
words=text.split()
words
```

```python
unique_words=set(words)
unique_words
```

```python
len(unique_words)
```

## Conclusion

Sets are a powerful and flexible data type in Python that provide a way to store collections of unique elements. They support various operations such as union, intersection, difference, and symmetric difference, which are useful for mathematical computations. Understanding how to use sets and their associated methods can help you write more efficient and clean Python code, especially when dealing with unique collections and membership tests.
