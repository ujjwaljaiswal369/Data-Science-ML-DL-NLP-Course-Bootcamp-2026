### Assignment 9: Working with the `datetime` Module

Use the `datetime` module to print the current date, calculate the date 100 days from today, and determine the day of the week for a given date.

```python
#solution-9

#importing_modules
from datetime import datetime,timedelta

#current_time
now=datetime.now()
print(f'Current Date : {now}')

#100_days_back_time
past_days = now - timedelta(days=100)
print(f'100 days back : {past_days}')

#Day_of_past_days
given_date=datetime(2026,3,1)
given_date.date()
print(f'''Day of 100 days back : {given_date.strftime('%A')}''')
```

### Assignment 10: Working with the `random` Module

Use the `random` module to generate a list of 5 random numbers between 1 and 50 and shuffle the elements of a list.

```python
#solution-10

import random

#Generating_random_numbers
random_no = [random.randint(1,50) for i in range(5)]
print(f'Random five numbers : {random_no}')

#Shuffling_the_list
random.shuffle(random_no)
print(f'Shuffled list : {random_no}')
```

### Assignment 11: Creating a Simple Package

Create a package named `mypackage` with two modules: `module1` and `module2`. `module1` should contain a function that adds two numbers, and `module2` should contain a function that multiplies two numbers. Write code to use these functions.

```python
#solution-11

#Importing_OS
import os

#Making_directory
try:
    os.mkdir('mypackage')
except FileExistsError :
    pass
    
#Going_inside_directory
os.chdir('mypackage')

#Creating_py_files
open('__init__.py','w').close()

with open('module1.py','w') as file:
    file.write('''
def addition(a,b):
    return a+b''')
    
with open('module2.py','w') as file:
    file.write('''
def multiply(a,b):
    return a*b''')
    
os.chdir('..')
os.getcwd()
```

```python
#testing
from mypackage import module1
from mypackage.subpackage import module2

print(module1.addition(1,2))
print(module2.multiply(2,4))
```
